```python
import pandas as pd

cpu_measures = pd.read_csv('cpu_measures.csv')
temp_measures = pd.read_csv('temp_measures.csv')
```

```python
cpu_measures_avg = cpu_measures.groupby('timestamp').mean()
cpu_measures_avg = cpu_measures_avg.drop('cpu', axis=1)

temp_measures_avg = temp_measures.groupby('timestamp').mean()

measures = pd.merge(cpu_measures_avg, temp_measures_avg, on='timestamp')
```

## All data

```python
measures.plot(figsize=(20,10))
```

```python
measures_with_time = measures.reset_index()
measures_with_time
```

## Fan - 25%, CPU usage with steps every 10%

```python
measures_filtered = measures_with_time[(measures_with_time['timestamp'] >= 1652477096) & (measures_with_time['timestamp'] <= 1652516275)]
measures_filtered.plot(x='timestamp', figsize=(20,10))
measures_filtered.to_csv('out_rpi_fan_25_steps10_baseline.csv', index=False)
```

## Fan - 25%, CPU usage with steps every 1%

```python
measures_filtered = measures_with_time[(measures_with_time['timestamp'] >= 1652518494) & (measures_with_time['timestamp'] <= 1652571310)]
measures_filtered.plot(x='timestamp', figsize=(20,10))
measures_filtered.to_csv('out_rpi_fan_25_steps1_baseline.csv', index=False)
```
